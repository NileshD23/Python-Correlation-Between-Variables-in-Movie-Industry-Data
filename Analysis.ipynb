```python
# Import libraries
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

# read in the data
df = pd.read_csv('movies.csv')

# view dataframe
df.head()
```

```python
# Check for missing values
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Drop missing values
df = df.dropna()

# drop duplicates as well
df.drop_duplicates()
```

```python
# Look at data types for our features
df.dtypes
```

```python
# Change budget and gross to an integer type
df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')

df.head()
```

```python
# Create correcte year column
df['year_correct'] = df['released'].str.extract(pat='([0-9]{4})').astype(int)

df
```

```python
# Check for outliers in gross
df.boxplot(column=['gross'])
```

```python
# Order dataframe by gross in descending order
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Plot budget vs gross using seaborn
sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color':'red'}, line_kws={'color':'blue'})
```

```python
# High correlation between budget and gross
```

```python
# Look at correlation for numeric features
df.corr(method='pearson') # could use kendall, spearman, etc.
```

```python
# Plot heatmap for numeric features
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title("Correlation Matrxi for Numeric Features")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")

plt.show()
```

```python
# Look at company
df.head()
```

```python
# Numerize the dataframe for the catergorical features
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
# Plot heatmap for categorical features
correlation_mat = df.corr(method='pearson')

sns.heatmap(correlation_mat, annot=True)

plt.title("Correlation Matrxi for Numeric Features")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")

plt.show()
```

```python
corr_mat = df_numerized.corr()

corr_pairs = corr_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
